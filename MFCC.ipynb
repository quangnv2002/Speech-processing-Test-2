```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_path = '/content/audio_data/c1.wav'
```

```python
ipd.Audio(audio_path)
```

```python
x, sr = librosa.load(audio_path)
```

```python
mfccs = librosa.feature.mfcc(x, sr=sr,n_mfcc=39)
```

```python
x.shape
```

```python
mfccs = librosa.feature.mfcc(x, n_mfcc=39, sr=sr)
```

```python
mfccs.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
delta_mfccs.shape
```

```python
delta2_mfccs.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta2_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```
